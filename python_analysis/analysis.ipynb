```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load backtest results
trades = pd.read_csv("results.csv")

# Compute cumulative PnL
trades['pnl'] = trades['price'] * trades['quantity']
trades['cum_pnl'] = trades['pnl'].cumsum()

plt.figure(figsize=(10,5))
plt.plot(trades['cum_pnl'])
plt.title("Cumulative PnL")
plt.xlabel("Trade Number")
plt.ylabel("PnL ($)")
plt.grid(True)
plt.show()

# Simple statistics
print("Total PnL:", trades['pnl'].sum())
print("Sharpe ratio (daily approx):", trades['pnl'].mean()/trades['pnl'].std()*np.sqrt(252))
```
